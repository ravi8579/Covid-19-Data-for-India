```python
infected = infected.groupby('Country/Region').sum() # Group by function applied to infected dataframe
recovered = recovered.groupby('Country/Region').sum() # Group by function applied to recovered dataframe
deaths = deaths.groupby('Country/Region').sum() # Group by function applied to deaths dataframe

infected.head()
```

```python
infected.loc['India'][2:].plot() # series for India location plotted from Infected dataframe
recovered.loc['India'][2:].plot() # series for India Location plotted from recovered dataframe
plt.show()# plot command
```

```python
infected.drop(columns=['Lat','Long'],inplace=True) # Longitude & Lattitude columns removed from infected dataframe
recovered.drop(columns=['Lat','Long'],inplace=True) # Longitude & Lattitude columns removed from recovered dataframe
deaths.drop(columns=['Lat','Long'],inplace=True) # Longitude & Lattitude columns removed from deaths dataframe
```

```python
#New Dataframe created with 3 columns infected,recovered,deaths from earlier daframes specific country rows & indexed with date time
# Dataframe created for India
def mkframe(country):
    df = pd.DataFrame({ 'infected' : infected.loc[country] ,
                        'recovered' : recovered.loc[country],
                        'deaths' : deaths.loc[country]})
    df.index = pd.to_datetime(df.index)
    return df

df = mkframe('India')
df
```

```python
df.plot() # new dataframe df plotted
plt.show()
```

```python
df['ninfected'] = df['infected'].diff() #New column ninfected created with diffrence command to find daily newly infected people i.e. daily increase in no of people compared previous days infected people
df['ninfected'].plot() # plotted newly infected people chart
plt.show()
```

```python
df[(df.index.year==2021) & (df.index.month==5)]['ninfected'].plot() # If we see this data in details there is disturbance in data due to reporting variation or delays.
plt.show()
```

```python
df['ninfav'] = df['ninfected'].rolling(window=8).mean() # So to smoothen the curve we have calculated rolling average 20 days to find mean newly infected people
df['ninfav'].plot()# rooling average plotted.
plt.show()
```

```python
countries[(countries['Country_Region']=='India') & countries['Province_State'].isna()] # Population for India from countries Dataframe
```

```python
# New series(Column) of Percentage of infected people in Country is found(India) New data frame Pinfected created from (infected people)*100 / population
pop = countries[(countries['Country_Region']=='India') & countries['Province_State'].isna()]['Population'].iloc[0]
df['pinfected'] = df['infected']*100 / pop
df['pinfected'].plot(figsize=(10,3))
plt.show()
```

```python
# Now new series of Rt is created as basic reproduction number Formulae
# As per Formulae Rt = (It-7+It-6+It-5+It-4)/(It-3+It-2+It-1+It) where 'It' is the newly infected people 'I' on 't' date. So 'It-7' means newly infected people on 7 days back from t date
df['Rt'] = df['ninfected'].rolling(8).apply(lambda x: x[4:].sum()/x[:4].sum())
df['Rt'].plot()
plt.show()
```

```python
#
ax = df[df.index<"2021-01-01"]['Rt'].replace(np.inf,np.nan).fillna(method='pad').plot(figsize=(20,3))
ax.set_ylim([0,16])
ax.axhline(1,linestyle='--',color='red')
plt.show()
```
